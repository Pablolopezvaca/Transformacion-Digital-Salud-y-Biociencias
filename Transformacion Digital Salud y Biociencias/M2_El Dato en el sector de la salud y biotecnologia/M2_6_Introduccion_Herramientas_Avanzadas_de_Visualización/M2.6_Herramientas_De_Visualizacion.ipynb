
# Subtema 6. Introducción a las Herramientas Avanzadas de Visualización

Las herramientas avanzadas de visualización son fundamentales para transformar datos complejos en representaciones gráficas que faciliten su comprensión y análisis. Estas herramientas permiten a los usuarios explorar, interpretar y comunicar datos de manera efectiva, revelando patrones, tendencias y relaciones que podrían pasar desapercibidas en tablas de datos crudas. Son esenciales en diversos campos, desde la investigación científica y la ingeniería hasta el marketing y las finanzas, proporcionando insights cruciales que apoyan la toma de decisiones informada.

## Definiciones y Ejemplos de Herramientas Avanzadas de Visualización

### 1. Gráficas de Líneas (Line Charts)

- Definición: Representan datos continuos a lo largo del tiempo con puntos conectados por líneas rectas.
- Ejemplo: Una gráfica de líneas puede usarse para mostrar la tendencia de las ventas diarias de una tienda a lo largo de un período de 10 días.

```python
import matplotlib.pyplot as plt

dias = list(range(1, 11))
ventas = [100, 150, 200, 250, 300, 350, 400, 450, 500, 550]

plt.plot(dias, ventas, marker='o', linestyle='-', color='b')
plt.title('Ventas Diarias')
plt.xlabel('Días')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

### 2. Gráficas de Barras (Bar Charts)

- Definición: Utilizan barras rectangulares para mostrar comparaciones entre categorías.
- Ejemplo: Una gráfica de barras puede comparar las ventas de diferentes categorías de productos.

```python
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]

plt.bar(categorias, valores, color='g')
plt.title('Comparación de Categorías')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

### 3. Histogramas

- Definición: Muestran la distribución de un conjunto de datos continuos dividiéndolos en intervalos (bins).
- Ejemplo: Un histograma puede representar la distribución de alturas en una población.

```python
import numpy as np

datos = np.random.randn(1000)
plt.hist(datos, bins=30, color='purple')
plt.title('Histograma de Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

### 4. Gráficas de Dispersión (Scatter Plots)

- Definición: Representan la relación entre dos variables mediante puntos dispersos en el plano.
- Ejemplo: Una gráfica de dispersión puede mostrar la relación entre la cantidad de horas estudiadas y las calificaciones obtenidas.

```python
x = np.random.rand(50)
y = x + np.random.rand(50) * 0.1

plt.scatter(x, y, color='red')
plt.title('Gráfica de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

### 5. Gráfico con varios ejes Y (Multiple Y-Axis Plot)

- Definición: Permiten mostrar dos o más series de datos con diferentes escalas de Y en el mismo gráfico.
- Ejemplo: Un gráfico con varios ejes Y puede mostrar simultáneamente la temperatura y la precipitación en un mismo gráfico.

```python
fig, ax1 = plt.subplots()

t = np.arange(0., 10., 0.2)
s1 = np.exp(t)
s2 = np.sin(t)

ax1.plot(t, s1, 'b-')
ax1.set_xlabel('Tiempo')
ax1.set_ylabel('Exp', color='b')

ax2 = ax1.twinx()
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('Sin', color='r')

plt.title('Gráfico con varios ejes Y')
plt.show()
```

### 6. Subplots

- Definición: Permiten incluir múltiples gráficos en una sola figura.
- Ejemplo: Los subplots se pueden usar para comparar la función seno y coseno en una sola visualización.

```python
fig, axs = plt.subplots(2)

x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

axs[0].plot(x, y1, 'b')
axs[0].set_title('Seno')

axs[1].plot(x, y2, 'r')
axs[1].set_title('Coseno')

for ax in axs.flat:
    ax.set(xlabel='x', ylabel='y')

plt.tight_layout()
plt.show()
```

### 7. Gráficas de Caja (Boxplots)

- Definición: Muestran la distribución de un conjunto de datos destacando los cuartiles y posibles valores atípicos.
- Ejemplo: Un boxplot puede utilizarse para comparar la distribución de salarios entre diferentes departamentos de una empresa.

```python
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(datos, vert=True, patch_artist=True)
plt.title('Gráficas de Caja')
plt.xlabel('Grupo')
plt.ylabel('Valor')
plt.show()
```

### 8. Gráficas de Violín (Violin Plots)

- Definición: Combinan aspectos de los boxplots y los histogramas para mostrar la distribución de los datos y su densidad.
- Ejemplo: Un violin plot puede mostrar la distribución de puntajes de exámenes entre diferentes clases.

```python
!pip install seaborn

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos de ejemplo
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Crear gráfica de violín
sns.violinplot(data=datos)
plt.title('Gráficas de Violín')
plt.xlabel('Grupo')
plt.ylabel('Valor')
plt.show()
```

### 9. Títulos, etiquetas y leyendas (Titles, Labels, and Legends)

- Definición: Elementos que describen el contenido del gráfico, las variables representadas y la información de los colores y estilos.
- Ejemplo: Añadir títulos, etiquetas y leyendas ayuda a interpretar correctamente un gráfico que compara dos series de datos.

```python
plt.plot([1, 2, 3], [4, 5, 6], label='Linea 1')
plt.plot([1, 2, 3], [6, 5, 4], label='Linea 2')

plt.title('Gráfico con Títulos, Etiquetas y Leyendas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

### 10. Colores y estilos de línea (Colors and Line Styles)

- Definición: Utilización de diferentes colores y estilos de línea para distinguir entre varias series de datos o para realzar aspectos específicos del gráfico.
- Ejemplo: Utilizar colores y estilos de línea diferentes para resaltar dos series de datos distintas en un mismo gráfico.

```python
plt.plot([1, 2, 3], [4, 5, 6], 'r--', label='Linea Roja Discontinua')
plt.plot([1, 2, 3], [6, 5, 4], 'b:', label='Linea Azul Punteada')

plt.title('Colores y Estilos de Línea')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

### 11. Anotaciones (Annotations)

- Definición: Textos o marcas adicionales en el gráfico para destacar puntos de interés o añadir contexto.
- Ejemplo: Anotar un punto máximo en una gráfica de funciones matemáticas para destacarlo.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.annotate('Máximo', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title('Gráfica con Anotaciones')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

### 12. Subplots de diferentes gráficos (Subplots of Different Charts)

- Definición: Incluir distintos tipos de gráficos en una sola figura para una comparación más completa.
- Ejemplo: Un subplot puede comparar una función seno y una exponencial en la misma figura.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.exp(x)

ax1.plot(x, y1, 'b')
ax1.set_title('Seno')

ax2.plot(x, y)
```

### 13. Gráficas 3D (3D Plots)

- Definición: Representaciones tridimensionales de datos que permiten visualizar relaciones entre tres variables.
- Ejemplo: Este ejemplo de código ilustra cómo crear una gráfica de dispersión en 3D para visualizar la relación entre tres variables. Este tipo de gráfica es útil en muchas áreas, como análisis de datos, visualización de datos científicos y más. En el contexto de la biociencia y la salud, podría utilizarse para visualizar la distribución de datos en tres dimensiones, como por ejemplo, la relación entre diferentes biomarcadores en estudios clínicos.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z)

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.title('Gráficas 3D')
plt.show()
```

### 1. Crear una gráfica de líneas con dos funciones diferentes y personalizarla. (Aquí puedes usar funciones trigonométricas o cualquier otra función matemática).

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos
x = np.linspace(0, 10, 100)  # 100 puntos entre 0 y 10
y1 = np.sin(x)  # Función seno
y2 = np.cos(x)  # Función coseno

# Crear la gráfica de líneas
plt.figure(figsize=(10, 6))  # Tamaño de la figura
plt.plot(x, y1, label='Seno', color='blue', linestyle='-', linewidth=2)  # Línea azul continua
plt.plot(x, y2, label='Coseno', color='red', linestyle='--', linewidth=2)  # Línea roja discontinua

# Personalizar la gráfica
plt.title('Gráfica de Líneas de Funciones Trigonométricas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()  # Mostrar la leyenda
plt.grid(True)  # Mostrar la cuadrícula

# Mostrar la gráfica
plt.show()
```

### Explicación del Ejemplo

En este ejemplo, hemos creado una gráfica de líneas que muestra dos funciones trigonométricas: seno y coseno. Cada función está representada por una línea diferente con estilos distintos para una mejor distinción visual. La gráfica también incluye una leyenda, etiquetas para los ejes, un título y una cuadrícula para mejorar su legibilidad y comprensión.

### 2. Crear una gráfica de barras con datos de ventas de productos.(Puedes simular datos de ventas para diferentes productos y mostrarlo en una gráfica de barras).

```python
import matplotlib.pyplot as plt

# Datos de ventas simulados
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E']
ventas = [150, 90, 200, 130, 170]

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))  # Tamaño de la figura
plt.bar(productos, ventas, color='skyblue')

# Personalizar la gráfica
plt.title('Ventas de Productos')
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.grid(axis='y')  # Mostrar la cuadrícula solo en el eje Y

# Mostrar los valores de ventas encima de las barras
for i, venta in enumerate(ventas):
    plt.text(i, venta + 5, str(venta), ha='center', va='bottom')

# Mostrar la gráfica
plt.show()
```

### Explicación del Ejemplo

Además, hemos añadido los valores de ventas encima de cada barra para que sea más fácil ver las cifras exactas. Esta gráfica es útil para visualizar y comparar rápidamente las ventas de diferentes productos.

### 4. Crear una gráfica de dispersión con datos aleatorios y personalizarla. (Incluye personalización de colores, tamaños de puntos y transparencias).

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
np.random.seed(0)  # Fijar la semilla para reproducibilidad
x = np.random.rand(100)
y = np.random.rand(100)
tamaños = 1000 * np.random.rand(100)  # Tamaños de los puntos
colores = np.random.rand(100)  # Colores de los puntos

# Crear la gráfica de dispersión
plt.figure(figsize=(10, 6))  # Tamaño de la figura
scatter = plt.scatter(x, y, s=tamaños, c=colores, alpha=0.6, cmap='viridis')

# Personalizar la gráfica
plt.title('Gráfica de Dispersión con Datos Aleatorios')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.colorbar(scatter, label='Color del Punto')  # Barra de colores

# Mostrar la gráfica
plt.show()
```

### Explicación del Ejemplo

En este ejemplo, hemos creado una gráfica de dispersión que visualiza 100 puntos generados aleatoriamente. Cada punto tiene un tamaño y un color aleatorios, lo que añade una dimensión adicional de información. La transparencia de los puntos (alpha=0.6) permite ver puntos superpuestos, y la colormap viridis proporciona una gradación de colores agradable. Esta gráfica es útil para visualizar la relación entre dos variables y cómo se distribuyen los datos en el espacio bidimensional, con información adicional proporcionada por los tamaños y colores de los puntos.

### 5. Crear un subplot con 4 gráficas diferentes. (Usa subplots para mostrar diferentes tipos de gráficas en una sola figura).

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos para las gráficas
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.exp(x)

# Crear la figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Gráfica 1: Línea
axs[0, 0].plot(x, y1, label='Seno', color='b')
axs[0, 0].set_title('Gráfica de Línea')
axs[0, 0].set_xlabel('Eje X')
axs[0, 0].set_ylabel('Eje Y')
axs[0, 0].legend()

# Gráfica 2: Dispersión
axs[0, 1].scatter(x, y2, label='Coseno', color='r')
axs[0, 1].set_title('Gráfica de Dispersión')
axs[0, 1].set_xlabel('Eje X')
axs[0, 1].set_ylabel('Eje Y')
axs[0, 1].legend()

# Gráfica 3: Barras
axs[1, 0].bar(np.arange(10), np.random.randint(1, 10, size=10), label='Barras', color='g')
axs[1, 0].set_title('Gráfica de Barras')
axs[1, 0].set_xlabel('Categorías')
axs[1, 0].set_ylabel('Valores')
axs[1, 0].legend()

# Gráfica 4: Línea con escala logarítmica
axs[1, 1].plot(x, y4, label='Exponencial', color='m')
axs[1, 1].set_yscale('log')
axs[1, 1].set_title('Gráfica de Línea con Escala Logarítmica')
axs[1, 1].set_xlabel('Eje X')
axs[1, 1].set_ylabel('Eje Y (log)')
axs[1, 1].legend()

# Ajustar el layout
plt.tight_layout()

# Mostrar la figura con subplots
plt.show()
```

### Explicación del Ejemplo

En este ejemplo, hemos creado una figura con 4 subplots diferentes:

- Gráfica de Línea: Muestra la función seno de x, útil para visualizar funciones matemáticas continuas.
- Gráfica de Dispersión: Muestra la función coseno de x con puntos dispersos, útil para ver la relación entre dos variables.
- Gráfica de Barras: Muestra valores de barras aleatorios, útil para comparar valores categóricos.
- Gráfica de Línea con Escala Logarítmica: Muestra la función exponencial de x en una escala logarítmica, útil para visualizar datos que crecen exponencialmente.

Este ejercicio demuestra cómo combinar diferentes tipos de gráficas en una sola figura utilizando subplots, permitiendo una comparación visual de diferentes conjuntos de datos o visualizaciones en una sola vista.

### 6. Crear una gráfica de caja con datos de diferentes categorías. (Usa datos generados aleatoriamente para diferentes categorías y muestra las cajas).

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios para diferentes categorías
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Crear la gráfica de caja
plt.figure(figsize=(10, 6))
plt.boxplot(data, patch_artist=True, notch=True, vert=True, showmeans=True)

# Personalizar la gráfica
plt.title('Gráfica de Caja')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.xticks([1, 2, 3], ['Categoría 1', 'Categoría 2', 'Categoría 3'])

# Mostrar la gráfica
plt.show()
```

### Explicación del Ejemplo

En este ejemplo, hemos creado una gráfica de caja con datos generados aleatoriamente para tres categorías diferentes. La gráfica de caja muestra la distribución de los datos y resalta la mediana, los cuartiles y los valores atípicos de cada categoría. Cada caja representa la distribución de los datos de una categoría específica.

Las características adicionales incluyen:

- Muescas en las cajas: Las muescas indican el intervalo de confianza alrededor de la mediana.
- Media de los datos: La media de los datos se muestra con un marcador especial.

Este tipo de gráfica es útil para comparar distribuciones de datos entre varias categorías y para identificar posibles valores atípicos.

### 7. Crear una gráfica de violín con datos generados aleatoriamente. (Muestra la distribución de datos con una gráfica de violín).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generar datos aleatorios para diferentes categorías
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Crear la gráfica de violín
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, inner="quartile", palette="muted")

# Personalizar la gráfica
plt.title('Gráfica de Violín')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.xticks([0, 1, 2], ['Categoría 1', 'Categoría 2', 'Categoría 3'])

# Mostrar la gráfica
plt.show()
```

### Explicación del Ejemplo

En este ejemplo, hemos creado una gráfica de violín con datos generados aleatoriamente para tres categorías diferentes. La gráfica de violín combina las características de un diagrama de caja y un gráfico de densidad de kernel, proporcionando una visualización de la distribución de los datos para cada categoría.

Las características adicionales incluyen:

- Cuartiles: Se muestran los cuartiles dentro de cada violín para indicar la dispersión de los datos.
- Paleta de colores: Se utiliza una paleta de colores suave para mejorar la visualización.

Este tipo de gráfica es útil para comparar distribuciones de datos entre varias categorías y para visualizar la densidad de los datos.

### 8. Ejercicio 8: Crear una gráfica 3D. (Usa datos generados y muestra una superficie 3D).

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generar datos para la superficie 3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Crear la figura y el eje 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Añadir una barra de color para indicar la escala
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Personalizar la gráfica
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.title('Superficie 3D')

# Mostrar la gráfica
plt.show()
```

### Explicación del Ejemplo

En este ejemplo, hemos creado una gráfica de superficie 3D utilizando datos generados aleatoriamente. La gráfica muestra una superficie tridimensional basada en una función sinusoidal de la distancia al origen (0,0). Esta visualización es útil para analizar la topografía y la forma de superficies complejas en un espacio tridimensional.

Las características adicionales incluyen:

- Barra de color: Se añade una barra de color para proporcionar una escala de colores que ayuda a interpretar los valores en la superficie.
- Personalización de ejes: Se añaden etiquetas a los ejes X, Y y Z, así como un título a la gráfica.

Esta gráfica es útil en varias áreas, incluyendo matemáticas, ingeniería y ciencias naturales, para visualizar superficies complejas y entender mejor la relación entre las variables.
