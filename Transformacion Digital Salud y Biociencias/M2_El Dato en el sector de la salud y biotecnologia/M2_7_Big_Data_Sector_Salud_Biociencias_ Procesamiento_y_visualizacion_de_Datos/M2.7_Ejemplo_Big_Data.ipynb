## Ejemplo de Big Data en el sector de la salud

Este ejemplo simula el manejo y procesamiento de un gran conjunto de datos de pacientes utilizando técnicas básicas de análisis y visualización. Generaremos datos de pacientes, incluyendo su ID, edad, peso, presión arterial y si tienen una enfermedad específica (diabetes).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulación de datos de pacientes
num_samples = 10**6  # Un millón de muestras para simular Big Data
np.random.seed(0)

# Generar datos aleatorios de pacientes
data = {
    'id': np.arange(num_samples),
    'age': np.random.randint(0, 100, size=num_samples),
    'weight': np.random.randint(50, 150, size=num_samples),  # Peso en kg
    'blood_pressure': np.random.randint(90, 180, size=num_samples),  # Presión arterial sistólica
    'diabetes': np.random.choice([0, 1], size=num_samples, p=[0.8, 0.2])  # 0: No, 1: Sí
}

df = pd.DataFrame(data)
df.head()
# Mostrar algunas estadísticas básicas
print(df.describe())

# Filtrar datos para encontrar pacientes con diabetes
diabetic_patients = df[df['diabetes'] == 1]
print(f"Number of diabetic patients: {len(diabetic_patients)}")
# Visualizar la distribución de edades
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribución de Edades de los Pacientes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
# Relación entre peso y presión arterial
plt.figure(figsize=(10, 6))
plt.scatter(df['weight'], df['blood_pressure'], alpha=0.5, s=1)
plt.title('Relación entre Peso y Presión Arterial')
plt.xlabel('Peso (kg)')
plt.ylabel('Presión Arterial (sistólica)')
plt.show()

```

## Explicación del ejemplo:

En este ejemplo, hemos simulado un conjunto de datos grande con un millón de registros que representan información de pacientes, incluyendo su ID, edad, peso, presión arterial y si tienen diabetes.

1. Importar las Librerías Necesarias: Utilizamos numpy para generar datos aleatorios, pandas para manejar y procesar los datos, y matplotlib para visualizar los datos.
2. Procesamiento de Datos: Calculamos estadísticas básicas y filtramos los datos para encontrar pacientes con diabetes.
3. Visualización de Datos: Creamos un histograma para mostrar la distribución de las edades de los pacientes.
4. Análisis de Datos: Visualizamos la relación entre el peso y la presión arterial utilizando una gráfica de dispersión.
